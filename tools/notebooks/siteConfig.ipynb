```typescript
import { logError } from "./logError.js";
import { reloadEnv } from "./reloadEnv.js";
await reloadEnv();

const eds_token = Deno.env.get("EDS_TOKEN");
const org = 'adobecom'
const site = 'upp'
```

```typescript
const response = await fetch(`https://admin.hlx.page/config/${org}/sites/${site}.json`, {
  headers: {
    'x-auth-token': eds_token,
  }
});

if (!response.ok) {
  await logError(response);
} else {
  const outputFile = `${site}.json`;
  const data = await response.json();
  await Deno.writeTextFile(outputFile, JSON.stringify(data, null, 2));
  console.log(data);
}
```
